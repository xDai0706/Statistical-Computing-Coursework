```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

import ps2
import ps3a
```

```python
def plot_mc_hand(hand, N, n):
    
    _, mc_scores = ps3a.play_n_mc_hand(hand, N, n)
    plt.hist(mc_scores, bins='auto')
    plt.title(f"Monte Carlo Scores for {n} rounds of {hand} with {N} samples")
    plt.xlabel("Score")
    plt.ylabel("Frequency")
    plt.show()
```

```python
hand = ps2.get_frequency_dict('helloworld')
random.seed(1)
for n in (10, 100):
    for N in (1, 10, 100):
        plot_mc_hand(hand, N, n)
```
